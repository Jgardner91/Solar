```python
import os
import time
import datetime
```

```python
def remove_cache():
    for i in range(100):
        location_1 = '/Users/jamesgardner/Library/Caches/Google/Chrome/Default/Cache/Cache_Data'
        cached = os.listdir('/Users/jamesgardner/Library/Caches/Google/Chrome/Default/Cache/Cache_Data')
        print(len(cached))
       
        for file in cached:
            if file != 'index-dir':
                path = os.path.join(location_1,file)
                os.remove(path)
        
        cached = os.listdir('/Users/jamesgardner/Library/Caches/Google/Chrome/Default/Cache/Cache_Data')
        print(len(cached))
        
        now = datetime.datetime.now()
        print(now.strftime("%d/%m/%Y %H:%M:%S"))
        time.sleep(300)
```
